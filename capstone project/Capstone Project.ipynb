```python
df1
```

```python
#Which City has the highest customers?
sns.countplot('Customer_type',hue='City', data=df1)
```

Naypyitaw had more member Customer_type, while Yangon had more normal customer_type

```python
#Which gender has the highest count in each of the Cities/Stores?
sns.countplot(x='City',hue='Gender', data=df1)
```

Naypyitaw and Mandalay had more Male , while Yangon had more Female

```python
#What product has the highest sales?
plt.figure(figsize=(10,4))
df1['Product_line'].value_counts().plot(kind='bar')
```

Fashion accessories had the highest sales

```python
#What product was bought most male?
plt.figure(figsize=(12,6))
sns.countplot('Product_line', hue= 'Gender',data = df1)
```

Health and beauty product was bought most by male, while Fashion accessories was bought most by Female  

```python
#Which gender gives the highest average rating?
sns.boxplot('Gender','Rating', data = df1)
```

Female gives the highest average rating
